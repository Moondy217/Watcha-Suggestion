```python
results = []

url = 'https://www.justwatch.com/kr/%EC%98%81%ED%99%94'
browser.get(url)

# OTT와 영화 수집 년도는 숫자 변경하여 따로 수집한 후 하나의 csv 파일로 합쳤음
# watcha, netflix
ott_name = 'netflix'

start = '2019'
end = '2019'

url = f'https://www.justwatch.com/kr/%EB%8F%99%EC%98%81%EC%83%81%EC%84%9C%EB%B9%84%EC%8A%A4/{ott_name}/%EC%98%81%ED%99%94%EC%82%B0%EC%97%85?release_year_from={start}&release_year_until={end}'

browser.get(url)
time.sleep(1) # 서버 부담 주지 않기 위해
    
scroll_to_bottom()

# BeautifulSoup을 사용하여 HTML 파싱
soup = BeautifulSoup(browser.page_source, "html.parser")
movie_list = soup.select('div.title-list-grid > div >  a')
```

```python
len(movie_list)
```

```python
rank = 0

for movie in movie_list:
        movie_link = movie['href']
        movie_url = f'https://www.justwatch.com{movie_link}'
        browser.get(movie_url)

        time.sleep(random.uniform(1,4))

        soup = BeautifulSoup(browser.page_source, "html.parser")
        
        # 예외 처리를 통해 요소 찾기 시도
        try:
            time.sleep(1)
            
            # 영화 이름 추출
            year = soup.select('div > span.text-muted')[0].text.strip()
            movie_name = soup.select('div > h1')[0].text.strip()
            
        except:
            try:
                soup = BeautifulSoup(browser.page_source, "html.parser")
                year = soup.select('div.jw-info-box.jw-info-box--no-card > div > div.jw-info-box__container-content > div:nth-child(2) > div.title-block__container > div.title-block > div > span')
                movie_name = soup.select('div.jw-info-box > div > div.jw-info-box__container-content > div:nth-child(2) > div.title-block__container > div.title-block > div > h1')[0].text.strip()
            
            except:
                soup = BeautifulSoup(browser.page_source, "html.parser")
                year = soup.select('div.jw-info-box > div > div.jw-info-box__container-content > div:nth-child(2) > div.title-block__container > div.title-block > div > span')
                
        movie_year = year.replace('(', '').replace(')', '')
            
        #영화 별점 추출 (IMDB)
        star = soup.select('div.title-info.visible-xs.visible-sm > div:nth-child(1) > div > div > div:nth-child(2) > span')

        if star:
            first_star = star[0]
            star_text = first_star.text.strip()  # '9.2 (1m)' 형태의 문자열
            rating = star_text.split()[0]  # 공백을 기준으로 분리하여 '9.2'만 추출

        
        rank = rank + 1
        print(rank, movie_year, movie_name, rating)

        data = [movie_year, movie_name, rating]  # 1개 채널 정보
        results.append(data)
```

```python
import pandas as pd
df = pd.DataFrame(results)
df.columns = ['개봉연도', '영화 제목', '별점']
df['별점'] = df['별점'].astype(float)
df = df.sort_values(by = '개봉연도', ascending = True)
df
```

```python
df.to_csv(f'./{ott_name}_{start}{end}.csv', index = False, encoding = 'utf-8-sig')
```
